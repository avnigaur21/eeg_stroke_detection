```python
import mne

stk= mne.io.read_raw_edf("sub-01_task-motor-imagery_eeg.edf", preload=True)
print(stk)
print(stk.ch_names)
```

```python
non = mne.io.read_raw_edf("S001R04.edf", preload=True)
print(non)
print(non.ch_names)
```

#### 4) Clean non-stroke channel names as they had extra dots and spaces and made them uppercase so that we can pick the common electrodes from both by comparing

```python
cleaned_channels = [ch.replace('.', '').upper() for ch in non.ch_names]
print(cleaned_channels)
```

#### 5) Now we will clean the stroke channels.
current stroke list has:

'HEOL', 'HEOR', ''

👉 These are: • EOG (eye movement)

```python
stroke_channels_clean = [ch for ch in stk.ch_names if ch not in ['HEOL', 'HEOR', '']]
print(stroke_channels_clean)
```

#### 6) Now we will standardize stroke channel names to match non-stroke.
Right now you have:

'Fz', 'Cz', 'Pz', 'Oz'

But in non-stroke:

'FZ', 'CZ', 'PZ', 'OZ'

👉 Case mismatch will cause problems later.

```python
stroke_channels_clean = [ch.upper() for ch in stroke_channels_clean]
print(stroke_channels_clean)
```

#### 7) Common channels between stroke and non-stroke
We want:

👉 Channels present in both datasets

👉 So data becomes comparable

```python
common_channels = list(set(stroke_channels_clean).intersection(set(cleaned_channels)))
print(sorted(common_channels))
```

#### 8) Now we will fix the order of channels.
👉 Right now your channels are unordered (random order from set)

👉 This can break:

• topographic maps ❌

• ML model consistency ❌

```python
ordered_channels = [ch for ch in stroke_channels_clean if ch in common_channels]
print(ordered_channels)
```

#### 9) Convert all channel names in the raw object to uppercase and pick using the uppercase list directly

```python
stk.rename_channels({ch: ch.upper() for ch in stk.ch_names})

stk.pick(ordered_channels)
print(stk.ch_names)
```

#### 10) Convert all channel names in the raw object to uppercase and remove trailing dots and pick using the uppercase list

```python
non.rename_channels({ch: ch.replace('.', '').upper() for ch in non.ch_names})

non.pick(ordered_channels)
print(non.ch_names)
```
